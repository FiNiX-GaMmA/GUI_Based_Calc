````python

from tkinter import *

from click import command



root = Tk()
root.title("GUI Calculator")
root.geometry("500x500")

# Function to add two numbers 
def add(num1, num2):
    return num1 + num2
  
# Function to subtract two numbers 
def subtract(num1, num2):
    return num1 - num2
  
# Function to multiply two numbers
def multiply(num1, num2):
    return num1 * num2
  
# Function to divide two numbers
def divide(num1, num2):
    return num1 / num2
  
def calculate():
    select = int(operationentry.get())
  
    number_1 = int(number1entry.get())
    number_2 = int(number2entry.get())
    if select == 1:
        Label(root, text = add(number_1, number_2)).place(x = 200, y = 300)
        
    elif select == 2:
        Label(root, text = subtract(number_1, number_2)).place(x = 200, y = 300)
       
    elif select == 3:
        Label(root, text = multiply(number_1, number_2)).place(x = 200, y = 300)
        
  
    elif select == 4:
        Label(root, text = divide(number_1, number_2)).place(x = 200, y = 300)
    
    else:
        Label(root, text = "Invalid operation").place(x = 200, y = 300)
    

Line0 = Label(root, text = "Please select an option").pack()
Line1 = Label(root, text = "1. Addition").pack()
Line2 = Label(root, text = "2. Subtraction").pack()
Line3 = Label(root, text = "3. Multiplication").pack()
Line4 = Label(root, text = "4. Division").pack()


number1 = Label(root, text = "Enter first number: ")
number1.place(x = 50, y = 110)
number1value = StringVar()
number1entry = Entry(root, textvariable = number1value)
number1entry.place(x = 200, y = 110)






number2 = Label(root, text = "Enter second number: ")
number2.place(x = 50, y = 160)
number2value = StringVar()
number2entry = Entry(root, textvariable = number2value)
number2entry.place(x = 200, y = 160)







operation = Label(root, text = "Enter operation: ")
operation.place(x = 50, y = 210)
operationvalue = StringVar()
operationentry = Entry(root, textvariable = operationvalue)
operationentry.place(x = 200, y = 210)



Button(root, text = "Calculate",command=calculate).place(x = 50, y = 260)
Button(root, text = "Exit",command=lambda:exit()).place(x = 150, y = 260)
Label(root, text = "Result = ").place(x = 150, y = 300)



root.mainloop()
````
